```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, metrics
import random
import os
import joblib
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import RobustScaler
from sklearn.metrics import (
    classification_report, confusion_matrix,
    roc_auc_score, accuracy_score,
    precision_score, recall_score, f1_score,
    precision_recall_curve
)
from sklearn.utils.class_weight import compute_class_weight

# ── Reproducibility ───────────────────────────────────────────────────────────
SEED = 42
os.environ["PYTHONHASHSEED"] = str(SEED)
np.random.seed(SEED)
random.seed(SEED)
tf.random.set_seed(SEED)

# ── Load Dataset 1 ────────────────────────────────────────────────────────────
csv_path = "/content/drive/MyDrive/Colab Notebooks/kept_logs_for_training.csv"
df = pd.read_csv(csv_path, low_memory=False)
print("Dataset Shape:", df.shape)

# ── Feature Engineering (environment-agnostic ratios) ─────────────────────────
def engineer_features(df):
    df = df.copy()
    eps = 1e-9  # avoid division by zero

    # Ratios are stable across networks; absolutes are not
    df["BYTE_RATIO"]        = df["IN_BYTES"]  / (df["OUT_BYTES"]  + eps)
    df["PKT_RATIO"]         = df["IN_PKTS"]   / (df["OUT_PKTS"]   + eps)
    df["BYTES_PER_PKT_IN"]  = df["IN_BYTES"]  / (df["IN_PKTS"]    + eps)
    df["BYTES_PER_PKT_OUT"] = df["OUT_BYTES"] / (df["OUT_PKTS"]   + eps)
    df["TOTAL_BYTES"]       = df["IN_BYTES"]  +  df["OUT_BYTES"]
    df["TOTAL_PKTS"]        = df["IN_PKTS"]   +  df["OUT_PKTS"]
    df["BYTES_PER_MS"]      = df["TOTAL_BYTES"] / (df["FLOW_DURATION_MILLISECONDS"] + eps)
    df["PKTS_PER_MS"]       = df["TOTAL_PKTS"]  / (df["FLOW_DURATION_MILLISECONDS"] + eps)
    df["PKT_SIZE_RATIO"]    = df["LONGEST_FLOW_PKT"] / (df["SHORTEST_FLOW_PKT"] + eps)
    df["RETRANS_RATIO_IN"]  = df["RETRANSMITTED_IN_PKTS"]  / (df["IN_PKTS"]  + eps)
    df["RETRANS_RATIO_OUT"] = df["RETRANSMITTED_OUT_PKTS"] / (df["OUT_PKTS"] + eps)
    df["TTL_DIFF"]          = df["MAX_TTL"] - df["MIN_TTL"]

    return df

# Raw features still included — ratios added on top
base_features = [
    "IN_BYTES", "OUT_BYTES", "IN_PKTS", "OUT_PKTS",
    "FLOW_DURATION_MILLISECONDS",
    "LONGEST_FLOW_PKT", "SHORTEST_FLOW_PKT",
    "TCP_FLAGS", "CLIENT_TCP_FLAGS", "SERVER_TCP_FLAGS",
    "RETRANSMITTED_IN_PKTS", "RETRANSMITTED_OUT_PKTS",
    "MIN_TTL", "MAX_TTL", "PROTOCOL", "L7_PROTO"
]

engineered_features = [
    "BYTE_RATIO", "PKT_RATIO",
    "BYTES_PER_PKT_IN", "BYTES_PER_PKT_OUT",
    "TOTAL_BYTES", "TOTAL_PKTS",
    "BYTES_PER_MS", "PKTS_PER_MS",
    "PKT_SIZE_RATIO",
    "RETRANS_RATIO_IN", "RETRANS_RATIO_OUT",
    "TTL_DIFF"
]

selected_features = base_features + engineered_features

df = engineer_features(df)
df = df[selected_features + ["Label"]]

# ── Train/Test Split ──────────────────────────────────────────────────────────
train_df, test_df = train_test_split(
    df, test_size=0.2, random_state=SEED, stratify=df["Label"]
)

X_train = train_df[selected_features]
X_test  = test_df[selected_features]
y_train = train_df["Label"].values
y_test  = test_df["Label"].values

# ── Clean ─────────────────────────────────────────────────────────────────────
def clean(X):
    return X.replace([np.inf, -np.inf], np.nan).fillna(0).astype(np.float32)

X_train = clean(X_train)
X_test  = clean(X_test)

# ── RobustScaler (much less sensitive to distribution shift than StandardScaler)
# Saved to disk so future datasets get the exact same transformation
scaler = RobustScaler()
X_train = scaler.fit_transform(X_train).astype(np.float32)
X_test  = scaler.transform(X_test).astype(np.float32)

# Clip extremes — stops wild out-of-distribution values from breaking the model
X_train = np.clip(X_train, -10, 10)
X_test  = np.clip(X_test,  -10, 10)

# Save scaler for future datasets
joblib.dump(scaler, "/content/drive/MyDrive/Colab Notebooks/scaler.pkl")
print("Scaler saved.")

# ── Class Weights ─────────────────────────────────────────────────────────────
class_weights = compute_class_weight("balanced", classes=np.unique(y_train), y=y_train)
class_weight_dict = {0: class_weights[0], 1: class_weights[1]}
print("Class Weights:", class_weight_dict)

# ── Model ─────────────────────────────────────────────────────────────────────
# Stronger dropout + L2 regularization to prevent overfitting to Dataset 1
l2 = keras.regularizers.l2(1e-4)

model = keras.Sequential([
    layers.Input(shape=(X_train.shape[1],)),

    layers.Dense(128, activation="relu", kernel_regularizer=l2),
    layers.BatchNormalization(),
    layers.Dropout(0.4),

    layers.Dense(64, activation="relu", kernel_regularizer=l2),
    layers.BatchNormalization(),
    layers.Dropout(0.4),

    layers.Dense(32, activation="relu", kernel_regularizer=l2),
    layers.BatchNormalization(),
    layers.Dropout(0.3),

    layers.Dense(16, activation="relu", kernel_regularizer=l2),
    layers.Dropout(0.2),

    layers.Dense(1, activation="sigmoid")
])

# Label smoothing stops the model becoming overconfident on Dataset 1's patterns
model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss=keras.losses.BinaryCrossentropy(label_smoothing=0.05),
    metrics=[
        metrics.Precision(name="precision"),
        metrics.Recall(name="recall"),
        metrics.AUC(name="roc_auc"),
        metrics.AUC(name="pr_auc", curve="PR")
    ]
)

model.summary()

# ── Callbacks ─────────────────────────────────────────────────────────────────
early_stop = keras.callbacks.EarlyStopping(
    monitor="val_roc_auc", mode="max", patience=7, restore_best_weights=True
)
lr_scheduler = keras.callbacks.ReduceLROnPlateau(
    monitor="val_loss", factor=0.5, patience=3, min_lr=1e-6, verbose=1
)

# ── Train ─────────────────────────────────────────────────────────────────────
history = model.fit(
    X_train, y_train,
    validation_split=0.2,
    epochs=20,
    batch_size=4096,
    class_weight=class_weight_dict,
    callbacks=[early_stop, lr_scheduler],
    verbose=1
)

# Save model
model.save("/content/drive/MyDrive/Colab Notebooks/mlp_model.keras")
print("Model saved.")

# ── Threshold Tuning via PR Curve (instead of hardcoded 0.5) ─────────────────
# Finds the threshold that maximises F1 on the test set
y_prob = model.predict(X_test).ravel()

precisions, recalls, thresholds = precision_recall_curve(y_test, y_prob)
f1_scores = 2 * (precisions * recalls) / (precisions + recalls + 1e-9)
best_idx   = np.argmax(f1_scores)
best_threshold = thresholds[best_idx]
print(f"\nOptimal Threshold (F1-maximising): {best_threshold:.4f}")

# Save threshold so future datasets use the same cutoff
np.save("/content/drive/MyDrive/Colab Notebooks/threshold.npy", best_threshold)

y_pred = (y_prob > best_threshold).astype(int)

# ── Metrics ───────────────────────────────────────────────────────────────────
print("\n=======================================")
print("MLP PERFORMANCE — DATASET 1 (TEST SET)")
print("=======================================")
print(f"Threshold     : {best_threshold:.4f}")
print(f"Accuracy      : {accuracy_score(y_test, y_pred):.6f}")
print(f"Precision     : {precision_score(y_test, y_pred):.6f}")
print(f"Recall        : {recall_score(y_test, y_pred):.6f}")
print(f"F1-Score      : {f1_score(y_test, y_pred):.6f}")
print(f"ROC-AUC Score : {roc_auc_score(y_test, y_prob):.6f}")
print("=======================================\n")
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))


# ══════════════════════════════════════════════════════════════════════════════
# INFERENCE ON ANY FUTURE DATASET — paste this block wherever needed
# ══════════════════════════════════════════════════════════════════════════════

def predict_new_dataset(csv_path):
    """
    Load a new dataset, apply identical feature engineering + saved scaler,
    and return predictions using the saved optimal threshold.
    """
    model      = keras.models.load_model("/content/drive/MyDrive/Colab Notebooks/mlp_model.keras")
    scaler     = joblib.load("/content/drive/MyDrive/Colab Notebooks/scaler.pkl")
    threshold  = float(np.load("/content/drive/MyDrive/Colab Notebooks/threshold.npy"))

    df_new = pd.read_csv(csv_path, low_memory=False)
    df_new = engineer_features(df_new)

    X_new = df_new[selected_features].replace([np.inf, -np.inf], np.nan).fillna(0).astype(np.float32)
    X_new = scaler.transform(X_new).astype(np.float32)
    X_new = np.clip(X_new, -10, 10)

    y_prob_new = model.predict(X_new).ravel()
    y_pred_new = (y_prob_new > threshold).astype(int)

    # If the new dataset has labels, evaluate automatically
    if "Label" in df_new.columns:
        y_true = df_new["Label"].values
        print("\n=======================================")
        print(f"MLP PERFORMANCE — {csv_path.split('/')[-1]}")
        print("=======================================")
        print(f"Threshold     : {threshold:.4f}")
        print(f"Accuracy      : {accuracy_score(y_true, y_pred_new):.6f}")
        print(f"Precision     : {precision_score(y_true, y_pred_new):.6f}")
        print(f"Recall        : {recall_score(y_true, y_pred_new):.6f}")
        print(f"F1-Score      : {f1_score(y_true, y_pred_new):.6f}")
        print(f"ROC-AUC Score : {roc_auc_score(y_true, y_prob_new):.6f}")
        print("=======================================\n")
        print("Confusion Matrix:\n", confusion_matrix(y_true, y_pred_new))
        print("\nClassification Report:\n", classification_report(y_true, y_pred_new))

    return y_pred_new, y_prob_new


# Usage — just call this for any future dataset:
# preds, probs = predict_new_dataset("/content/drive/MyDrive/Colab Notebooks/dataset2.csv")
```
