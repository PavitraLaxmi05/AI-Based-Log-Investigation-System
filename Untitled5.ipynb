```python
import random
import time
```

```python
def generate_synthetic_dna_bitstring(length=8):
    bitstring = "".join(random.choice("01") for _ in range(length))
    return bitstring

def simulate_qml_team_output(batches=5):
    print(f"--- Q-Nexus Synthetic DNA Stream Started ---")
    generated_data = []
    for i in range(batches):
        time.sleep(1)
        dna_sample = generate_synthetic_dna_bitstring(8)
        generated_data.append(dna_sample)

        print(f"[Batch {i+1}]: Generated Synthetic Sequence: {dna_sample}")

    print(f"--- Generation Complete. {batches} samples ready for PQC Vault ---")
    return generated_data
if __name__ == "__main__":
    samples = simulate_qml_team_output(10)
    print(f"Final Data for Blockchain: {samples}")
```

```python
import hashlib
import json
```

```python
class QNexusBlock:
    def __init__(self, index, dna_data, previous_hash):
        self.index = index
        self.timestamp = time.time()
        self.dna_data = dna_data
        self.previous_hash = previous_hash
        self.hash = self.calculate_quantum_safe_hash()
    def calculate_quantum_safe_hash(self):
        block_content = {"index": self.index, "timestamp": self.timestamp, "dna_data": self.dna_data, "previous_hash": self.previous_hash}
        block_string = json.dumps(block_content, sort_keys=True).encode()
        return hashlib.sha3_512(block_string).hexdigest()

class QNexusLedger:
    def __init__(self):
        self.chain = [self.create_genesis_block()]
    def create_genesis_block(self):
        return QNexusBlock(0, "GENESIS", "0"*128)

    def add_dna_record(self, dna_bitstring):
        prev_block = self.chain[-1]
        new_block = QNexusBlock(len(self.chain), dna_bitstring, prev_block.hash)
        self.chain.append(new_block)
        print(f"Block #{new_block.index} Secured. Hash: {new_block.hash[:16]}...")

    def is_chain_valid(self):
        for i in range(1, len(self.chain)):
            current = self.chain[i]
            previous = self.chain[i-1]
            if current.hash != current.calculate_quantum_safe_hash():
                return False
            if current.previous_hash != previous.hash:
                return False
        return True
if __name__ == "__main__":
    my_vault = QNexusLedger()
    print("Starting...")
    new_dna_samples = [generate_synthetic_dna_bitstring(8) for _ in range(5)]
    for sample in new_dna_samples:
        my_vault.add_dna_record(sample)
        time.sleep(0.5)
    print(f"\nBlockchain Integrity Verified: {my_vault.is_chain_valid()}")
```

```python
import secrets

class QNexusSigner:
    def __init__(self):
        self.private_key = secrets.token_hex(64)

    def sign_data(self, data):
        signature_base = data + self.private_key
        return hashlib.sha3_512(signature_base.encode()).hexdigest()
hospital_signer = QNexusSigner()

def secure_add_with_signature(ledger, dna_sample, signature, signer_public_check):
    """
    This stops the hacker before the block is even created!
    """
    # Verify signature (Simulation)
    expected_sig = hospital_signer.sign_data(dna_sample)

    if signature == expected_sig:
        ledger.add_dna_record(dna_sample)
        print("✅ Access Granted: Signature Verified.")
    else:
        print("❌ ACCESS DENIED: Invalid Quantum Signature. Block Rejected!")

# --- TEST IT ---
sample_dna = "10101100"
valid_sig = hospital_signer.sign_data(sample_dna)

print("\n--- Testing Prevention Layer ---")
# Unauthorized attempt (Hacker doesn't have the private key)
secure_add_with_signature(my_vault, sample_dna, "fake_signature_123", None)

# Authorized attempt
secure_add_with_signature(my_vault, sample_dna, valid_sig, None)
```
